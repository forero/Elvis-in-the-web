```python
#==================================================================================================
#Importing all necessary libraries
#==================================================================================================
%pylab inline
from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# JSAnimation import available at https://github.com/jakevdp/JSAnimation
from JSAnimation import IPython_display
from matplotlib import animation
import matplotlib.image as mpimg
import numpy as np
import os
#Customized cmap
my_cmap4 = plt.cm.get_cmap('gray', 4)
```

```python
#==================================================================================================
#Parameters
#==================================================================================================
#Web scheme to use
web = "Tweb"
#Resolution
res = 128
#Threshold value
l_thr = 0.5
#Box Lenght
Lbox = 50.
#Gaussian softening lenght
lsoft = 2.0

#data folder
datafold = "../../data/%s%d/"%(web, res)
#Eigenvalue name
eigname = lambda box, eig: "%s/GVD_F09_1001_000_z125_box%d_009.s%1.2f.eigen_%d"%(datafold,box,lsoft,eig)
eignamenull = lambda box: "%s/GVD_F09_1001_000_z125_box%d_009.s%1.2f.eigen"%(datafold,box,lsoft)

#Indices of Boxes
boxes = [1,3,4,5,7,8,14,15,17,20]
Nboxes = len(boxes)
```

```python
#==================================================================================================
#General functions
#==================================================================================================
#Classification Scheme
def Scheme( eig1, eig2, eig3, Lamb ):
    N1 = len( eig1 )
    try:
        N2 = len( eig1[0] )
        sch = 0*eig1
        for i in xrange(N1):
            for j in xrange(N2):
                if eig1[i,j] > Lamb: sch[i,j] += 1
                if eig2[i,j] > Lamb: sch[i,j] += 1
                if eig3[i,j] > Lamb: sch[i,j] += 1
        return sch
    except:
        sch = 0*eig1
        for i in xrange(N1):
            if eig1[i] > Lamb: sch[i] += 1
            if eig2[i] > Lamb: sch[i] += 1
            if eig3[i] > Lamb: sch[i] += 1
        return sch
        
#Cut Field
def CutFieldZ( filename, X, res=32, Coor = 3, N=64 ):
    '''
    Coor  1 -- X      2 -- Y      3 -- Z
    '''
    if res == 'plain':
	os.system( "./Field_Cut_Plain.out %s %d temp.tmp %d %d"%( filename, X, Coor, N ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
      
    else:	
	os.system( "./Field_Cut%d.out %s %d temp.tmp %d"%( res, filename, X, Coor ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
    return datos

#Fractional anisotropy Function
def Fractional_Anisotropy( eig1, eig2, eig3 ):
    FA = 1/np.sqrt(3.)*np.sqrt( ( (eig1 - eig3)**2 + (eig2 - eig3)**2 + (eig1 - eig2)**2 )/ \
    (eig1**2 + eig2**2 + eig3**2))
    return FA

#FA anisotropy correlator
def FA_lambda1_correlation( filename_eig, lamb_min, lamb_max, N_lam, FA_min, FA_max, N_FA ):
    os.system( "./FA_L1_Correlation.out %s %f %f %d %f %f %d"%
    ( filename_eig, lamb_min, lamb_max, N_lam, FA_min, FA_max, N_FA ) )
    datos = np.loadtxt( "temp.tmp" )
    os.system( "rm temp.tmp" )
    return datos
```

# Tweb visual impression

[Contents](#Contents)

```python
#Plotting environment
def VisualWeb( box, axe = 3, coor = 0, axes = None ):
    #Loading axis
    if axis == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Loading Fields
    eig1 = CutFieldZ( eigname(box,1), coor, 16, Coor = axe )
    eig2 = CutFieldZ( eigname(box,2), coor, 16, Coor = axe )
    eig3 = CutFieldZ( eigname(box,3), coor, 16, Coor = axe )
    
    #Plotting
    axes.imshow( -Scheme( eig1, eig2, eig3, l_thr )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=-3, vmax=0,
    cmap = my_cmap4, interpolation=None )
    axes.set_xlabel( "X [$h^{-1}\ Mpc$]", fontsize=14 )
    axes.set_ylabel( "Y [$h^{-1}\ Mpc$]", fontsize=14 )
    axes.set_title("Tweb for box %d"%(box), fontsize=14 )
    axes.legend( )
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
fig.subplots_adjust( hspace = 0.1, wspace=0.4 )
axes = []
for i in xrange(0,Nboxes):
    axes.append( fig.add_subplot(2,5,i+1) )
    VisualWeb( boxes[i], axes=axes[i] )
```

```python
#Plotting FA field
def VisualFA( box, axe = 3, coor = 0, axes = None ):
    #Loading axis
    if axis == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Loading Fields
    eig1 = CutFieldZ( eigname(box,1), coor, 16, Coor = axe )
    eig2 = CutFieldZ( eigname(box,2), coor, 16, Coor = axe )
    eig3 = CutFieldZ( eigname(box,3), coor, 16, Coor = axe )
    
    #Plotting
    axes.imshow( Fractional_Anisotropy( eig1, eig2, eig3 )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=0, vmax=1,
    cmap = "hot_r", interpolation=None )
    axes.set_xlabel( "X [$h^{-1}\ Mpc$]", fontsize=14 )
    axes.set_ylabel( "Y [$h^{-1}\ Mpc$]", fontsize=14 )
    axes.set_title("Tweb for box %d"%(box), fontsize=14 )
    axes.legend( )
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
fig.subplots_adjust( hspace = 0.1, wspace=0.4 )
axes = []
for i in xrange(0,Nboxes):
    axes.append( fig.add_subplot(2,5,i+1) )
    VisualFA( boxes[i], axes=axes[i] )
```

# FA-$\lambda_1$ correlations

```python
#FA-L1 correlator
def FA_L1_correlator( box, L_min = -0.5, L_max = 1.5, N_lamb = 100, FA_min = 0.0, FA_max = 1.0, N_FA = 100, axes = None ):
    #Loading axes
    if axes == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Arrays
    Lambda = np.linspace( L_min, L_max, N_lamb )
    FA = np.linspace( FA_min, FA_max, N_FA )
    
    #Building the matrix
    FA_L1 = FA_lambda1_correlation( eignamenull(box), L_min, L_max, N_lamb, FA_min, FA_max, N_FA )
    
    #Initializing counters
    Quartiles = np.zeros( (N_lamb,4) )
    for i_l in xrange( N_lamb ):
          FA_L1[i_l,:] = np.cumsum( FA_L1[i_l,:] )
          FA_L1[i_l,:] /= FA_L1[i_l,-1]
          median = True
          Q1 = True
          Q3 = True
          Quartiles[i_l,0] = Lambda[i_l]
          for i_fa in xrange( N_FA ):
              #Finding Median
              if FA_L1[i_l,i_fa] > 0.5 and median:
                  median = False
                  Quartiles[i_l,2] = FA[i_fa]
              #Finding Q1
              if FA_L1[i_l,i_fa] > 0.25 and Q1:
                  Q1 = False
                  Quartiles[i_l,1] = FA[i_fa]
              #Finding Q3
              if FA_L1[i_l,i_fa] > 0.75 and Q3:
                  Q3 = False
                  Quartiles[i_l,3] = FA[i_fa]
        
    axes.fill_between( Lambda, Quartiles[:,1], Quartiles[:,3], alpha = 0.3 )
    axes.plot( Lambda, Quartiles[:,2], "-", lw=2.0 )
    axes.set_xlim( (L_min,L_max) )
    axes.set_ylim( (FA_min,FA_max) )
    axes.hlines( 0.95, L_min, L_max, color="black", lw=2.0, zorder=20 )
    axes.set_xlabel( "$\lambda_1$" )
    axes.set_ylabel( "$FA$" )
    axes.set_title( "FA-$\lambda_1$" )
    axes.grid(1)
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)
for i in xrange(0,Nboxes):
    FA_L1_correlator( boxes[i], axes=axes1 )
    FA_L1_correlator( boxes[i], FA_min = 0.8, axes=axes2 )
```

According to this figure, the optim lambda threshold for all the boxes is $\lambda_{th} = 0.5$

```python
#==================================================================================================
#Importing all necessary libraries
#==================================================================================================
%pylab inline
from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# JSAnimation import available at https://github.com/jakevdp/JSAnimation
from JSAnimation import IPython_display
from matplotlib import animation
import matplotlib.image as mpimg
from matplotlib.colors import colorConverter
import numpy as np
import os

#Customized cmap
my_cmap4 = plt.cm.get_cmap('gray', 4)
#Customize 2map
def cmap2(color):
    cmap2b = mpl.colors.LinearSegmentedColormap.from_list('my_cmap2',["white",color],2)
    cmap2b._init()
    alphas = np.ones( cmap2b.N+3)
    alphas[0] = 0
    cmap2b._lut[:,-1] = alphas
    return cmap2b

#Label of axes
xlabels = ["Y", "X", "X"]
xindex = [2,1,1]
ylabels = ["Z", "Z", "Y"]
yindex = [3,3,2]
zlabels = ["X", "Y", "Z"]
#Colors of neighbour
colors = ["blue", "green", "red", "cyan", "purple"]
```

```python
#==================================================================================================
#Parameters
#==================================================================================================
#Web scheme to use
web = "Tweb"
#Resolution
res = 64
#Threshold value
l_thr = 0.8
#Box Lenght
Lbox = 50.
#Gaussian softening lenght
lsoft = 1.0

#data folder
datafold = "../../data/%s%d/"%(web, res)
#Eigenvalue name
eigname = lambda box, eig: "%s/GVD_F09_1001_000_z125_box%d_009.s%1.2f.eigen_%d"%(datafold,box,lsoft,eig)
eignamenull = lambda box: "%s/GVD_F09_1001_000_z125_box%d_009.s%1.2f.eigen"%(datafold,box,lsoft)

#Indices of Boxes
boxes = [1,3,4,5,7,8,14,15,17,20]
Nboxes = len(boxes)
```

```python
#==================================================================================================
#General functions
#==================================================================================================
#Classification Scheme
def Scheme( eig1, eig2, eig3, Lamb ):
    N1 = len( eig1 )
    try:
        N2 = len( eig1[0] )
        sch = 0*eig1
        for i in xrange(N1):
            for j in xrange(N2):
                if eig1[i,j] > Lamb: sch[i,j] += 1
                if eig2[i,j] > Lamb: sch[i,j] += 1
                if eig3[i,j] > Lamb: sch[i,j] += 1
        return sch
    except:
        sch = 0*eig1
        for i in xrange(N1):
            if eig1[i] > Lamb: sch[i] += 1
            if eig2[i] > Lamb: sch[i] += 1
            if eig3[i] > Lamb: sch[i] += 1
        return sch
        
#Cut Field
def CutFieldZ( filename, X, res=32, Coor = 3, N=64 ):
    '''
    Coor  1 -- Z      2 -- Y      3 -- X
    '''
    if res == 'plain':
	os.system( "./Field_Cut_Plain.out %s %d temp.tmp %d %d"%( filename, X, Coor, N ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
      
    else:	
	os.system( "./Field_Cut%d.out %s %d temp.tmp %d"%( res, filename, X, Coor ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
    return datos

#Fractional anisotropy Function
def Fractional_Anisotropy( eig1, eig2, eig3 ):
    FA = 1/np.sqrt(3.)*np.sqrt( ( (eig1 - eig3)**2 + (eig2 - eig3)**2 + (eig1 - eig2)**2 )/ \
    (eig1**2 + eig2**2 + eig3**2))
    return FA

#FA anisotropy correlator
def FA_lambda1_correlation( filename_eig, lamb_min, lamb_max, N_lam, FA_min, FA_max, N_FA ):
    os.system( "./FA_L1_Correlation.out %s %f %f %d %f %f %d"%
    ( filename_eig, lamb_min, lamb_max, N_lam, FA_min, FA_max, N_FA ) )
    datos = np.loadtxt( "temp.tmp" )
    os.system( "rm temp.tmp" )
    return datos
```

# Tweb visual impression

[Contents](#Contents)

```python
#Plotting environment
def VisualWeb( box, axe = 3, coor = 0, axes = None ):
    #Loading axis
    if axis == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Loading Fields
    eig1 = CutFieldZ( eigname(box,1), coor, 16, Coor = axe )
    eig2 = CutFieldZ( eigname(box,2), coor, 16, Coor = axe )
    eig3 = CutFieldZ( eigname(box,3), coor, 16, Coor = axe )
    
    #Plotting
    axes.imshow( -Scheme( eig1, eig2, eig3, l_thr )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=-3, vmax=0,
    cmap = my_cmap4, interpolation=None )
    axes.set_xlabel( "%s [$h^{-1}\ Mpc$]"%(xlabels[axe-1]), fontsize=14 )
    axes.set_ylabel( "%s [$h^{-1}\ Mpc$]"%(ylabels[axe-1]), fontsize=14 )
    axes.set_title("Tweb for box %d"%(box), fontsize=14 )
    axes.legend( )
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
fig.subplots_adjust( hspace = 0.1, wspace=0.4 )
axes = []
for i in xrange(0,Nboxes):
    axes.append( fig.add_subplot(2,5,i+1) )
    VisualWeb( boxes[i], axes=axes[i] )
```

```python
#Plotting FA field
def VisualFA( box, axe = 3, coor = 0, axes = None ):
    #Loading axis
    if axis == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Loading Fields
    eig1 = CutFieldZ( eigname(box,1), coor, 16, Coor = axe )
    eig2 = CutFieldZ( eigname(box,2), coor, 16, Coor = axe )
    eig3 = CutFieldZ( eigname(box,3), coor, 16, Coor = axe )
    
    #Plotting
    axes.imshow( Fractional_Anisotropy( eig1, eig2, eig3 )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=0, vmax=1,
    cmap = "hot_r", interpolation=None )
    axes.set_xlabel( "%s [$h^{-1}\ Mpc$]"%(xlabels[axe-1]), fontsize=14 )
    axes.set_ylabel( "%s [$h^{-1}\ Mpc$]"%(ylabels[axe-1]), fontsize=14 )
    axes.set_title("Tweb for box %d"%(box), fontsize=14 )
    axes.legend( )
    return axes
```

# FA-$\lambda_1$ correlations

[Contents](#Contents)

```python
#FA-L1 correlator
def FA_L1_correlator( box, L_min = -0.5, L_max = 1.5, N_lamb = 100, FA_min = 0.0, FA_max = 1.0, N_FA = 100, axes = None ):
    #Loading axes
    if axes == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Arrays
    Lambda = np.linspace( L_min, L_max, N_lamb )
    FA = np.linspace( FA_min, FA_max, N_FA )
    
    #Building the matrix
    FA_L1 = FA_lambda1_correlation( eignamenull(box), L_min, L_max, N_lamb, FA_min, FA_max, N_FA )
    
    #Initializing counters
    Quartiles = np.zeros( (N_lamb,4) )
    for i_l in xrange( N_lamb ):
          FA_L1[i_l,:] = np.cumsum( FA_L1[i_l,:] )
          FA_L1[i_l,:] /= FA_L1[i_l,-1]
          median = True
          Q1 = True
          Q3 = True
          Quartiles[i_l,0] = Lambda[i_l]
          for i_fa in xrange( N_FA ):
              #Finding Median
              if FA_L1[i_l,i_fa] > 0.5 and median:
                  median = False
                  Quartiles[i_l,2] = FA[i_fa]
              #Finding Q1
              if FA_L1[i_l,i_fa] > 0.25 and Q1:
                  Q1 = False
                  Quartiles[i_l,1] = FA[i_fa]
              #Finding Q3
              if FA_L1[i_l,i_fa] > 0.75 and Q3:
                  Q3 = False
                  Quartiles[i_l,3] = FA[i_fa]
        
    axes.fill_between( Lambda, Quartiles[:,1], Quartiles[:,3], alpha = 0.3 )
    axes.plot( Lambda, Quartiles[:,2], "-", lw=2.0 )
    axes.set_xlim( (L_min,L_max) )
    axes.set_ylim( (FA_min,FA_max) )
    axes.hlines( 0.95, L_min, L_max, color="black", lw=2.0, zorder=20 )
    axes.set_xlabel( "$\lambda_1$" )
    axes.set_ylabel( "$FA$" )
    axes.set_title( "FA-$\lambda_1$" )
    axes.grid(1)
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)
for i in xrange(0,Nboxes):
    FA_L1_correlator( boxes[i], axes=axes1 )
    FA_L1_correlator( boxes[i], FA_min = 0.8, axes=axes2 )
```

According to this figure, the optim lambda threshold for all the boxes is $\lambda_{th} = 0.5$

# Visual impression of Voids

[Contents](#Contents)

```python
#Plotting voids with colors
def VisualVoids( box, axe = 3, coor = 0, axes = None ):
    #Loading axes
    if axes == None:
        fig = plt.figure()
        axes = fig.add_subplot(1,1,1)
        
    #Catalogue of voids
    voids = CutFieldZ( "%s/voids/voids%d/void_index.dat"%(datafold, box), coor, 'plain', Coor=axe, N=res )
    
    #Calculating contour of voids
    contours = np.zeros( (res,res) )
    #Sweeping matrix
    for i in xrange(res):
        for j in xrange(res):
            for ic in arange(-1,2):
                for jc in arange(-1,2):
                    #Indexes of neighbours
                    it = i + ic
                    jt = j + jc
                    if i + ic >= res: it = 0
                    if i + ic < 0: it = res-1
                    if j + jc >= res: jt = 0
                    if j + jc < 0: jt = res-1
                    #Marking contours
                    if( voids[i,j] != voids[it,jt] ):
                        contours[i,j] = 1
    #Void basins
    num_voids = np.max( voids )
    lista = np.array([nan] + list(np.random.permutation( range(1,num_voids.astype(int)+1) )))
    #voids
    voids2 = lista[ voids.astype(int) ]
    axes.imshow( np.transpose(voids2[::,::-1]), cmap = 'spectral', extent = [0,Lbox,0,Lbox], 
    vmin = -num_voids/50., vmax = num_voids, interpolation = 'none' )
    axes.imshow( np.transpose(contours[::,::-1]), cmap = cmap2(color='black'), extent = [0,Lbox,0,Lbox], 
    vmin = 0, vmax = 1 )
    
    #Format
    axes.set_xlabel( "%s [$h^{-1}\ Mpc$]"%(xlabels[axe-1]), fontsize=14 )
    axes.set_ylabel( "%s [$h^{-1}\ Mpc$]"%(ylabels[axe-1]), fontsize=14 )
    axes.set_title("Voids box %d"%(box), fontsize=14 )
    axes.legend( )
    
    return axes
```

```python
fig = plt.figure( figsize=(16,6.5) )
fig.subplots_adjust( hspace = 0.1, wspace=0.4 )
axes = []
for i in xrange(0,Nboxes):
    axes.append( fig.add_subplot(2,5,i+1) )
    VisualWeb( boxes[i], axes=axes[i] )
```

```python
fig = plt.figure( figsize=(16,6.5) )
fig.subplots_adjust( hspace = 0.1, wspace=0.4 )
axes = []
for i in xrange(0,Nboxes):
    axes.append( fig.add_subplot(2,5,i+1) )
    VisualWeb( boxes[i], axes=axes[i] )
    VisualVoids( boxes[i], axes=axes[i] )
```

# Environment of LG systems

[Contents](#Contents)

```python
#halo_environment
def halo_environment( box, idh, dx=None, Nneigh=1 ):
    #Loading halos
    catalog = np.loadtxt( "%s/voids/halos%d.dat"%(datafold, box), usecols=(0,1,2,3) )
    #Id of nearest voids
    idvoid = np.loadtxt( "%s/voids/distances%d.dat"%(datafold, box) )
    #Loading discrete indexes of the halo
    ih = int(np.floor(catalog[idh,1]*res/Lbox))
    jh = int(np.floor(catalog[idh,2]*res/Lbox))
    kh = int(np.floor(catalog[idh,3]*res/Lbox))
    inh = [ih,jh,kh]

    #Plotting projections in each plane
    fig = plt.figure( figsize=(15,6) )
    
    #Plane XY============================================================================================
    for index in xrange(3):
        subplot = fig.add_subplot(1,3,index+1)
        #Indexes of the other coordinates
        i1 = xindex[index]
        j1 = yindex[index]
        
        #Plotting current halo
        subplot.plot( [catalog[idh,i1],], [catalog[idh,j1],], "o", color="red", ms=6 )
                
        #Plotting environment of LG
        VisualWeb( box, axe = index+1, coor = inh[index], axes=subplot )
        
        #Catalogue of voids
        voids = CutFieldZ( "%s/voids/voids%d/void_index.dat"%(datafold, box), inh[index], 'plain', Coor=index+1, N=res )
        #Plotting the closer neighbours
        for ingb in xrange(Nneigh):
            #Filtering no belonging voids
            void = voids==idvoid[idh,5*ingb+1]
            #Plotting voids
            subplot.imshow( np.transpose(void[::,::-1]), extent = [0,Lbox,0,Lbox], interpolation = 'none',\
            cmap=cmap2(colors[ingb]) )

        #Format
        subplot.set_title( "Halo %d box%d (%s = %1.2f)"%\
        (catalog[idh,0], box, zlabels[index],catalog[idh,index+1]), fontsize=14 )
        
        if dx != None:
            subplot.set_xlim( (catalog[idh,i1]-dx,catalog[idh,i1]+dx) )
            subplot.set_ylim( (catalog[idh,j1]-dx,catalog[idh,j1]+dx) )
        else:
            subplot.set_xlim( (0,Lbox) )
            subplot.set_ylim( (0,Lbox) )
            
    return None
```
